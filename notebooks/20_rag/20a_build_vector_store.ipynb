# 20a — Build Vector Store

**Goal:** Chunk docs, embed, and index into FAISS.

**Inputs:** Local markdown files

**Outputs:** FAISS index

> Tip: Fill your API keys in `.env` (copy from `.env.example`). Never commit secrets.


```python
import os, glob
docs = []
for p in glob.glob("../../docs/**/*.md", recursive=True):
    with open(p, "r", encoding="utf-8") as f:
        docs.append({"path": p, "text": f.read()})
print(f"Loaded {len(docs)} docs for indexing (stub).")
```
