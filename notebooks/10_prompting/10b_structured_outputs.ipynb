# 10b — Structured Outputs

**Goal:** Constrain outputs to JSON via schema + validation.

**Inputs:** Instruction + schema

**Outputs:** Validated JSON

> Tip: Fill your API keys in `.env` (copy from `.env.example`). Never commit secrets.


```python
from pydantic import BaseModel, ValidationError

class Recipe(BaseModel):
    title: str
    ingredients: list[str]
    steps: list[str]

candidate = {"title": "Tea", "ingredients": ["water","tea"], "steps": ["boil","steep"]}
print(Recipe(**candidate).model_dump_json(indent=2))
```
